```python
# Import libraries
import dataset2020
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from datetime import datetime
```

```python
def dataInMonthRange(csv_filepath, startMonth, endMonth):
    """
    Input: csv_filepath (a csv or excel file), startMonth (an integer between 1 and 12), endMonth (an integer between 1 and 12)
    Output:  a Pandas dataframe with csv data with DateTime stamp between startMonth and endMonth (inclusive)
    """
    df = dataset2020.cleaned_df(csv_filepath) # Import MyLA311 data
    
    # Filter data from startMonth to endMonth
    df['month'] = pd.DatetimeIndex(df['CreatedDate']).month # Create month column with value 1-12 representing CreatedDate month
    
    is_after_startMonth = df['month'] >= startMonth 
    is_before_endMonth = df['month'] <= endMonth
    
    df_in_range = df[is_after_startMonth & is_before_endMonth] 
    return df_in_range
```

```python
# Import and run datasets though dataInMonthRange
df_2017 = dataInMonthRange(r"C:\Users\hanaa\Downloads\MyLA311_Service_Request_Data_2017.csv", 3, 10)
df_2018 = dataInMonthRange(r"C:\Users\hanaa\Downloads\MyLA311_Service_Request_Data_2018.csv", 3, 10)
df_2019 = dataInMonthRange(r"C:\Users\hanaa\Downloads\MyLA311_Service_Request_Data_2019.csv", 3, 10)
df_2020 = dataInMonthRange(r"C:\Users\hanaa\Downloads\MyLA311_Service_Request_Data_2020.csv", 3, 10)
```

```python
def convertToDateTime(df):
    df['CreatedDate'] = pd.DatetimeIndex(df['CreatedDate']) # Change CreatedDate row to Datetime object
    #print(df.dtypes) # Check that CreatedDate is datatime type
    return df.head()

for df in [df_2017, df_2018, df_2019, df_2020]:
    convertToDateTime(df)
```
