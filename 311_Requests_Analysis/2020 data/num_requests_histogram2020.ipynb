```python
# Import libraries
import dataset2020 #Our custom code with the pipeline, elapsed time function, etc.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from datetime import datetime
import seaborn as sns
from scipy import stats
```

```python
df = dataset2020.cleaned_df('/Users/antoniorevilla/Downloads/cleaned_MyLA311_Service_Request_Data_2020.csv')
```

# Separating data into March-October (2020)

```python
df['month'] = pd.DatetimeIndex(df['CreatedDate']).month # Create month column with value 1-12 representing CreatedDate month
df['month']
```

```python
df1 = df[(df['month'] >= 3) & (df['month'] <= 10)]# Limits dataset to months >= 3 (March) and months <= 10 (October)
df1['month']
```

```python
num_bins = len(df1['month'].unique())# Limits dataset to months >= 3 (March) and months <= 10['month'].unique()) # Set bin number to the # of months (i.e. the number of unique values in the month col )
```

# Plotting Histograms (Matplotlib)

```python
# Matplotlib Historgram for 2020 Requests

# Set plot size
plt.figure(figsize=(12,7))

# Add title and axis names
plt.title('March-October 2020')
plt.xlabel('month')
plt.ylabel('# of requests')

# Limits for the X axis -- start to end month
plt.xlim(3,10) 

plt.hist(df1['month'], bins=num_bins, edgecolor='black')
```

# Plotting Histograms(Pandas)

```python

# Pandas Histogram for 2020 Requests

# Add title and axis names
plt.title('March-October 2020')
plt.xlabel('month')
plt.ylabel('# of requests')

# Set x limits -- start to end month
plt.xlim(3,10) 

hist = df1['month'].hist(bins=num_bins, figsize=(12,7), edgecolor='Black')
```
